```python
import pandas as pd
Covid_19 = "csse_covid_19_daily_reports/04-02-2020.csv"
# To add headers as we load the data...
df = pd.read_csv(Covid_19)
```

```python
#Total confirmed deaths , recovered by country
df2=pd.pivot_table(df,
 index=['Country_Region'],
 aggfunc='sum',
 values=['Deaths','Recovered','Confirmed'])
df2.sort_values(by='Deaths',ascending=False)
```

```python
df2["Deaths"].plot(title="Death by COVID19")
```
